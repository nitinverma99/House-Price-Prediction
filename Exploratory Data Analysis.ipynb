```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

```python
dataset.shape
```

# Missing Values

```python
# Here we will not deal with the missing values but check the 
# percentage of nan values present in each feature

# Step-1 : Make the list of features that have missing values
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

# Step-2 : Print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4)) 
```

```python
# It is seen that there a lot of missing values, so we need to find the relationship between 
# missing values and Sales Price

for feature in features_with_na:
    data = dataset.copy()
    
    # Make a variable that indicates 1 if the observation if missing or zero
    data[feature] = np.where(data[feature].isnull(),1,0)
    # if null, then 1 otherwise as 0
    
    # Now, calculate the mean SalesPrice where data is missing or present
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
# With the relation with between the missing values and the dependent variable is clearly visible.
# So, we need to replace these nan values with something meaningful which will be done in feature engineering.
```

# Numerical Variables


```python
# List 
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# Content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Analyzing the Temporal Datetime Variables
# Checking the relationship between the year the house is sold and the SalesPrice

dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs YearSold")
```

```python
year_feature
```

```python
# The house price is declining
# So, we will compare the difference between all years feature with SalesPrice

for feature in year_feature:
    if feature != "YrSold":
        data = dataset.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+["Id"]]
print("Count", len(discrete_feature))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# Relationship with the salesPrice

plt.style.use("ggplot")
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

```python
## Continuous Variable

continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+["Id"]]
print("Count", len(continuous_feature))
```

```python
# We analyze the continuous variables using histogram

for feature in continuous_feature:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```
